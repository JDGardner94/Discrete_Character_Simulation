```python
#Create a Clade size dictionary with # lists = # runs
CladeSize = {}
for i in range(1,Run+1):
    CladeSize.update({"Run" + str(i):[]})

#print(CladeSize)
```

```python
#Create Rate dictionaries with # lists = # runs
DarwinRateDictq12 = {}
DarwinRateDictq13 = {}
DarwinRateDictq21 = {}
DarwinRateDictq24 = {}
DarwinRateDictq31 = {}
DarwinRateDictq34 = {}
DarwinRateDictq42 = {}
DarwinRateDictq43 = {}

for i in range(1,Run+1):
    DarwinRateDictq12.update({"q12_"+str(i):[]})
    DarwinRateDictq13.update({"q13_"+str(i):[]})
    DarwinRateDictq21.update({"q21_"+str(i):[]})
    DarwinRateDictq24.update({"q24_"+str(i):[]})
    DarwinRateDictq31.update({"q31_"+str(i):[]})
    DarwinRateDictq34.update({"q34_"+str(i):[]})
    DarwinRateDictq42.update({"q42_"+str(i):[]})
    DarwinRateDictq43.update({"q43_"+str(i):[]})
```

```python
#Create dictionaries for the root state probabilities
DarwinRoot00 = {}
DarwinRoot01 = {}
DarwinRoot10 = {}
DarwinRoot11 = {}

for i in range(1,Run+1):
    DarwinRoot00.update({"Run_"+str(i):[]})
    DarwinRoot01.update({"Run_"+str(i):[]})
    DarwinRoot10.update({"Run_"+str(i):[]})
    DarwinRoot11.update({"Run_"+str(i):[]})
```

```python
#Create dictionary for every log file
DarwinIt = []
```

```python
for i in range(1,2):
    DarwinFilename = DarwinMacpd + "darwin_1." + str(i) + ".txt.Log.txt"
    DarwinOpenfile=open(DarwinFilename,'r')
    print(DarwinFilename)
    DarwinLines = DarwinOpenfile.readlines()[49:]
    for line in DarwinLines:
        DarwinIt.append(int(line.split('\t')[0]))
```

```python
for i in range(1,Run+1):
    DarwinFilename = DarwinMacpd + "darwin_1." + str(i) + ".txt.log.txt"
    DarwinOpenfile=open(DarwinFilename,'r')
    print(DarwinFilename)
    DarwinLines = DarwinOpenfile.readlines()[49:]
    for line in DarwinLines:
        DarwinLh["Lh" + str(i)].append(float(line.split('\t')[1]))
        DarwinDepInd["DI" + str(i)].append(str(line.split('\t')[6]))
        DarwinRateDictq12["q12_"+str(i)].append(float(line.split('\t')[7]))
        DarwinRateDictq13["q13_"+str(i)].append(float(line.split('\t')[8]))
        DarwinRateDictq21["q21_"+str(i)].append(float(line.split('\t')[9]))
        DarwinRateDictq24["q24_"+str(i)].append(float(line.split('\t')[10]))
        DarwinRateDictq31["q31_"+str(i)].append(float(line.split('\t')[11]))
        DarwinRateDictq34["q34_"+str(i)].append(float(line.split('\t')[12]))
        DarwinRateDictq42["q42_"+str(i)].append(float(line.split('\t')[13]))
        DarwinRateDictq43["q43_"+str(i)].append(float(line.split('\t')[14]))
        DarwinRoot00["Run_"+str(i)].append(float(line.split('\t')[15]))
        DarwinRoot01["Run_"+str(i)].append(float(line.split('\t')[16]))
        DarwinRoot10["Run_"+str(i)].append(float(line.split('\t')[17]))
        DarwinRoot11["Run_"+str(i)].append(float(line.split('\t')[18]))
```

```python
for i in range(1,Run+1):
    DataFilename = DarwinMacpd + "darwin_1." + str(i) + ".txt"
    DataOpenfile=open(DataFilename,'r')
    print(DataFilename)
    DataLines = DataOpenfile.readlines()
    for line in DataLines:
        CladeSize["Run_" + str(i)].append(float(line.split('\t')[2]))
```
